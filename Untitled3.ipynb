```python
class Upsampling(tf.keras.layers.Layer):
    def __init__(self,num_ker, name=None):
        super(Upsampling, self).__init__(name=name)

        self.convT = tf.keras.layers.Conv2DTranspose(num_ker, 3, strides=(2,2), 
                                                 padding='same', output_padding=(1,1),
                                                 kernel_initializer="he_uniform",)
        self.batchNorm = tf.keras.layers.BatchNormalization(gamma_initializer='ones', beta_initializer='zeros')
        self.act = tf.keras.layers.ReLU()

    def __call__(self, x):
        x = self.convT(x)
        x = self.batchNorm(x)
        x = self.act(x)
        return x
```

```python
class Generator(tf.keras.layers.Layer):
    def __init__(self, n_class, name=None):
        super(Generator, self).__init__(name=name)

        #self.ca = ConditioningAugmentation()
        self.embed = tf.keras.layers.Embedding(n_class, 256)
        self.den1 = tf.keras.layers.Dense(4*4)
        self.resh1 = tf.keras.layers.Reshape((4, 4, 1))
        self.den2 = tf.keras.layers.Dense(1024*4*4)
        self.act1 = tf.keras.layers.LeakyReLU(alpha=0.2)
        self.resh2 = tf.keras.layers.Reshape((4, 4, 1024))
        self.concat = tf.keras.layers.Concatenate(axis=3)

        self.up1 = Upsampling(512)
        self.up2 = Upsampling(256)
        self.up3 = Upsampling(128)
        self.up4 = Upsampling(64)
        self.conv = tf.keras.layers.Conv2D(3, kernel_size=3,
                                           padding='same', strides=1,
                                           kernel_initializer='he_uniform')
        self.actTanH = tf.keras.layers.Activation(tf.keras.activations.relu)
    def __call__(self, x):
        #x = self.ca(x)
        x = self.embed(x)
        x = self.den1(x)
        x = self.resh1(x)
        noise = tf.keras.layers.Input(shape=(100,))
        noise = self.den2(noise)
        noise = self.act1(noise)
        noise = self.resh2(noise)
        x = self.concat([x, noise])
        x = self.up1(x)
        x = self.up2(x)
        x = self.up3(x)
        x = self.up4(x)
        x = self.conv(x)
        x = self.actTanH(x)

        return x
```

```python
class Discriminator(tf.keras.layers.Layer):
    def __init__(self, n_class, name=None):
        super(Discriminator, self).__init__(name=name)

        self.embed = tf.keras.layers.Embedding(n_class, 256)
        self.den1 = tf.keras.layers.Dense(64*64)
        self.resh = tf.keras.layers.Reshape((64, 64, 1))
        self.concat = tf.keras.layers.Concatenate(axis=3)

        self.down1 = Downsampling(128)
        self.down2 = Downsampling(256)
        self.down3 = Downsampling(512)
        self.down4 = Downsampling(512)
        self.flat = tf.keras.layers.Flatten()
        self.drop = tf.keras.layers.Dropout(0.4)
        self.den2 = tf.keras.layers.Dense(1, activation='sigmoid')

    def __call__(self, input):
        x = input[0]
        lab = input[1]
        lab = self.embed(lab)
        lab = self.den1(lab)
        lab = self.resh(lab)
        x = self.concat([x, lab])
        x = self.down1(x)
        x = self.down2(x)
        x = self.down3(x)
        x = self.down4(x)
        x = self.flat(x)
        x = self.drop(x)
        x = self.den2(x)
        return x
```

```python
class Downsampling(tf.keras.layers.Layer):
    def __init__(self,num_ker, name=None):
        super(Downsampling, self).__init__(name=name)

        self.conv = tf.keras.layers.Conv2D(num_ker, 3, strides=(2,2), 
                                           padding='same', kernel_initializer="he_uniform",)
        self.batchNorm = tf.keras.layers.BatchNormalization(gamma_initializer='ones', beta_initializer='zeros')
        self.act = tf.keras.layers.LeakyReLU(alpha=0.2)

    def __call__(self, x):
        x = self.conv(x)
        x = self.batchNorm(x)
        x = self.act(x)
        return x
```

```python
def Upsampling_func(x, num_ker):
  x = tf.keras.layers.Conv2DTranspose(num_ker, 3, strides=(2,2), 
                                                 padding='same', output_padding=(1,1),
                                                 kernel_initializer="he_uniform",)(x)
  x = tf.keras.layers.BatchNormalization(gamma_initializer='ones', beta_initializer='zeros')(x)
  x = tf.keras.layers.ReLU()(x)
  return x
```

```python
def Generator_func(num_class):
  input_label = tf.keras.layers.Input(shape=(1,))
  lab = tf.keras.layers.Embedding(num_class, 256)(input_label)
  lab = tf.keras.layers.Dense(4*4)(lab)
  lab = tf.keras.layers.Reshape((4, 4, 1))(lab)

  noise = tf.keras.layers.Input(shape=(100,))
  noi = tf.keras.layers.Dense(1024*4*4)(noise)
  noi = tf.keras.layers.LeakyReLU(alpha=0.2)(noi)
  noi = tf.keras.layers.Reshape((4, 4, 1024))(noi)

  mer = tf.keras.layers.Concatenate()([lab, noi])

  mer = Upsampling_func(mer, 512)
  mer = Upsampling_func(mer, 256)
  mer = Upsampling_func(mer, 128)
  mer = Upsampling_func(mer, 64)

  mer = tf.keras.layers.Conv2D(3, kernel_size=3,
                                           padding='same', strides=1,
                                           kernel_initializer='he_uniform')(mer)
  out = tf.keras.layers.Activation(tf.keras.activations.relu)(mer)

  model_func = tf.keras.models.Model([input_label, noise], out)
  return model_func

```

```python
def Downsampling_func(x, num_ker):
  x = tf.keras.layers.Conv2D(num_ker, 3, strides=(2,2), 
                                           padding='same', kernel_initializer="he_uniform",)(x)
  x = tf.keras.layers.BatchNormalization(gamma_initializer='ones', beta_initializer='zeros')(x)
  x = tf.keras.layers.LeakyReLU(alpha=0.2)(x)
  return x
```

```python
def Discriminator_func(num_class):
  input_label = tf.keras.layers.Input(shape=(1,))
  lab = tf.keras.layers.Embedding(num_class, 256)(input_label)
  lab = tf.keras.layers.Dense(64*64)(lab)
  lab = tf.keras.layers.Reshape((64, 64, 1))(lab)

  img_label = tf.keras.layers.Input(shape=(64, 64, 3,))

  mer = tf.keras.layers.Concatenate()([img_label, lab])

  mer = Downsampling_func(mer, 128)
  mer = Downsampling_func(mer, 256)
  mer = Downsampling_func(mer, 512)
  mer = Downsampling_func(mer, 512)

  mer = tf.keras.layers.Flatten()(mer)
  mer = tf.keras.layers.Dropout(0.4)(mer)
  out = tf.keras.layers.Dense(1, activation='sigmoid')(mer)

  model_func = tf.keras.models.Model([img_label, input_label], out)
  optimizeAdam = tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)
  model_func.compile(loss='binary_crossentropy', optimizer=optimizeAdam, metrics=['accuracy'])

  return model_func

  
```

```python
model_dis = Discriminator_func(40)
print(model_dis.summary())
#tf.keras.utils.plot_model(model_dis, to_file='/content/drive/MyDrive/Comp4211/discriminator_model.png')
```

```python
def GAN_func(gen, dis):
  dis.trainable = False

  g_label, g_noise = gen.input
  g_out = gen.output

  d_out = dis([g_out, g_label])

  model_func = tf.keras.models.Model([g_noise, g_label], d_out)
  optimizeAdam = tf.keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)
  model_func.compile(loss='binary_crossentropy', optimizer=optimizeAdam)
  
  return model_func
```

```python
class CondGan():
  def __init__(self,num_class, dataset, epoch, num_batch):
    self.Generator = Generator_func(num_class)
    print(self.Generator.summary())
    self.Discriminator = Discriminator_func(num_class)
    print(self.Discriminator.summary())
    self.CondGan = GAN_func(self.Generator, self.Discriminator)
    print(self.CondGan.summary())
    self.num_class = num_class
    self.dataset = dataset
    self.epoch = epoch
    self.num_batch = num_batch
    self.half_batch = int(num_batch/2)
    self.epochBatch = int(dataset[0].shape[0] / num_batch)
    self.halfNumBatch = int(self.num_batch/2)
  def train(self):
    for i in range(self.epoch):
      for j in range(self.epochBatch):
        realIndex = np.random.randint(0, self.dataset[0].shape[0], self.half_batch)
        realImg, realLabels = self.dataset[0][realIndex], self.dataset[1][realIndex]
        realY = np.ones((self.half_batch, 1))

        dis_loss_real, _ = self.Discriminator.train_on_batch([realImg, realLabels], realY)

        n_dist = np.random.randn(100 * self.half_batch)
        noise_input = n_dist.reshape(self.half_batch, 100)
        fakeLabels = np.random.randint(0, self.num_class, self.half_batch).reshape(-1,1)
        FakeImg = self.Generator.predict([fakeLabels, noise_input])
        fakeY = np.zeros((self.half_batch, 1))

        dis_loss_fake, _ = self.Discriminator.train_on_batch([FakeImg, fakeLabels], fakeY)

        n_dist_gan = np.random.randn(100 * self.num_batch)
        noise_gan_input = n_dist_gan.reshape(self.num_batch, 100)
        fakeLabels_gan = np.random.randint(0, self.num_class, self.num_batch)

        ganY = np.ones((self.num_batch, 1))

        gan_loss = self.CondGan.train_on_batch([noise_gan_input, fakeLabels_gan], ganY)

        print('Epoch>%d, Batch%d/%d, d1=%f, d2=%f g=%f' %
              (i+1, j+1, self.epochBatch, dis_loss_real, dis_loss_fake, gan_loss))

    self.Generator.save('trafficSigns_generator.h5')





```
